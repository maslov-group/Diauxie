```python
fig, ax = plt.subplots(figsize = (5, 5))
x = np.linspace(0, 1.2, 200)
y = x
ax.plot(x, y, linestyle = '-.', color = 'k')
ax.fill_between(x, y, color = 'g', alpha = 0.3)
ax.fill_betweenx(x, y, color = 'y', alpha = 0.3)
ax.plot([0, 0.4], [0.4, 0.4], color = '#63a0cb')
ax.plot([0.4, 0.8], [0.4, 0], color = '#63a0cb')
ax.scatter(0.65, 0.15, marker = 'o', color = 'b', s = 150, label = 'steady state', zorder = 10)
ax.set_xlim(0, 1.2)
ax.set_ylim(0, 1.2)
ax.set_xlabel('$T_2$')
ax.set_ylabel('$T_1$')
ax.set_xticks([0.8])
ax.set_xticklabels(['log$D/g_1^{(2)}$'])
ax.set_yticks([0.4])
ax.set_yticklabels(['log$D/g_1^{(1)}$'])
#plt.legend()
plt.savefig('figsversion1/fig4-1.pdf', bbox_inches = 'tight')
```
